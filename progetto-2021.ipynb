### 2. For each book in the *Books* dataset, compute its average rating.

Tramite la funzione _merge_ uniamo i dataset __books__ e __books_ratings__, successivamente raggruppiamo per Titolo e ISBN e facciamo la media. Infine, per rendere più chiaro il dataset, rinominiamo la colonna con le medie.

```python
books_avg_rating=books.merge(books_ratings, how='left', on='ISBN').groupby(['Book-Title','ISBN'],as_index=False)[['Book-Rating']].mean()
books_avg_rating.rename(columns = {'Book-Rating':'Avg-Rating'}, inplace = True)
books_avg_rating.head()
```

### 4. Merge together all rows sharing the same book title, author and publisher. We will call the resulting dataset `merged books`. The books that have not been merged together will not appear in `merged books`.

Inizialmente raggruppiamo le righe per Titolo, Autore e Editore, e contiamo, tramite la funzione _count_, quante istanze condividono gli stessi valori in questi 3 campi. Generiamo così un nuovo dataset. Manteniamo nel dataset creato solo i valori che si ripetono più di una volta, così facendo eliminiamo i libri che non sono stati uniti. Infine resettiamo l'indice per agevolare la lettura del dataset.

```python
merged_books = books.groupby(['Book-Title','Book-Author','Publisher'],as_index=False).count()
merged_books = merged_books[merged_books['ISBN']>1][['Book-Title','Book-Author','Publisher']]
merged_books.reset_index(inplace=True, drop=True)
merged_books.head()
```

### 5. For each book in `merged books` compute its average rating.

The average is computed considering all books in `books` that have been merged.

Inizialmente uniamo, tramite la funzione _merge_, il dataset __books__ e il dataset __books_ratings__, raggruppando le righe per Titolo, Autore ed Editore e calcolandone la media. Inseriamo il risultato di queste operazioni in un nuovo dataset chiamato __books_ratings_merged__.

Successivamente uniamo il dataset appena creato con il dataset creato al punto 4, __merged_books__, avendo così, per ogni libro in __merged_books__, la media dei voti.

```python
books_ratings_merged = books.merge(books_ratings, how='left', on='ISBN').groupby(['Book-Title','Book-Author','Publisher'], as_index=False)['Book-Rating'].mean()
books_ratings_merged.rename(columns = {'Book-Rating':'Avg-Book-Rating'}, inplace = True)
merged_books_avg = merged_books.merge(books_ratings_merged, how='left', on=['Book-Title','Book-Author','Publisher'])
merged_books_avg.head()
```

### 6. For each book in `merged books` compute the minimum and maximum of the average ratings over all corresponding books in the `books` dataset.

Hence for each book in `merged books` we will have exactly two values (a minimum and a maximum)

Inizialmente uniamo, tramite la funzione _merge_, il dataset __books__ e il dataset __books_ratings__, raggruppando le righe per Titolo, Autore, Editore e ISBN e calcolandone la media. In questo caso abbiamo raggruppato anche per ISBN, allo scopo di calcolare la media per ogni edizione di un libro (contraddistinta dall'ISBN). Inseriamo il risultato di queste operazioni in un nuovo dataset chiamato __books_ratings_merged_ISBN__.

Successivamente uniamo il dataset appena creato con il dataset creato al punto 4, __merged_books__, avendo così, per ogni libro in __merged_books__, la media dei voti per ogni ISBN. Creiamo poi due colonne, __Max-Rating__ e __Min-Rating__, nelle quali inseriamo, per ogni combinazione di Titolo, Autore ed Editore, il voto minimo e il voto massimo calcolato sull'ISBN.

```python
books_ratings_merged_ISBN = books.merge(books_ratings, how='left', on='ISBN').groupby(['Book-Title','Book-Author','Publisher','ISBN'], as_index=False)['Book-Rating'].mean()
merged_books_avg_ISBN = merged_books.merge(books_ratings_merged_ISBN, on=['Book-Title','Book-Author','Publisher'])
merged_books['Max-Rating'] = merged_books_avg_ISBN.groupby(['Book-Title','Book-Author','Publisher'], as_index=False).max()['Book-Rating']
merged_books['Min-Rating'] = merged_books_avg_ISBN.groupby(['Book-Title','Book-Author','Publisher'], as_index=False).min()['Book-Rating']
merged_books.head()
```

### 7. For each book in `goodbooks`, compute the list of its authors. Assuming that the number of reviews with a text (column `work_text_reviews_count`) is split equally among all authors, find for each authors the total number of reviews with a text. We will call this quantity the *shared number of reviews with a text*.

Prima di tutto, tramite la funzione _split_, suddividiamo la colonna __authors__ ad ogni virgola. Generiamo così una colonna, __authors_list__, contenente per ogni libro un vettore di stringhe con i nomi degli autori.

Utilizziamo la lunghezza dei vettori generati, ottenuta tramite _apply(len)_, per calcolare il numero di recensioni per ogni autore (__shared_number_of_reviews_with_a_text__).

Successivamente generiamo una riga per ogni autore tramite _explode_, con lo scopo di poter raggruppare le righe per autore e sommare il numero di recensioni per ogni autore.

```python
goodbooks['authors_list'] = goodbooks['authors'].str.split(',')
goodbooks['shared_number_of_reviews_with_a_text'] = goodbooks['work_text_reviews_count']/goodbooks['authors_list'].apply(len)
goodbooks_authors_exploded = goodbooks.explode('authors_list',ignore_index=True)
goodbooks_authors_exploded.groupby(['authors_list'], as_index=False).sum('shared_number_of_reviews_with_a_text')[['authors_list','shared_number_of_reviews_with_a_text']].head(37)
```

Dall'output notiamo che l'ordine alfabetico ricomincia alla linea 35, un motivo di questo potrebbe essere che nell'utilizzo della funzione _split_ siano rimasti gli spazi antecedenti i termini successivi al primo in ogni lista.

Verifichiamo quindi questa ipotesi.

```python
goodbooks_authors_exploded['authors_list'][2]
```

Attraverso la funzione _strip_ eliminiamo gli spazi in eccesso.

```python
goodbooks_authors_exploded['authors_list'] = goodbooks_authors_exploded['authors_list'].str.strip()
goodbooks_authors_exploded.groupby(['authors_list'], as_index=False).sum('shared_number_of_reviews_with_a_text')[['authors_list','shared_number_of_reviews_with_a_text']].head()
```

### 8. For each year of publication, determine the author that has the largest value of the shared number of reviews with a text.

Per ogni anno e autore calcoliamo la somma delle recensioni. Successivamente per ogni anno troviamo l'autore con più recensioni.

```python
goodbooks_authors_sum_per_year = goodbooks_authors_exploded.groupby(['original_publication_year','authors_list']).sum()
goodbooks_authors_sum_per_year.loc[goodbooks_authors_sum_per_year.groupby(['original_publication_year']).idxmax()['shared_number_of_reviews_with_a_text']]['shared_number_of_reviews_with_a_text'].head()
```

### 9. Assuming that there are no errors in the ISBN fields, find the books in both datasets, and compute the difference of average rating according to the ratings and the goodratings datasets

Dopo aver unito il dataset __goodbooks__ con il dataset __books_avg_rating__, creato al punto 2, calcoliamo la differenza tra i voti medi dei due dataset (__average_rating__ e __Avg-Rating__). Osserviamo che la scala di valutazione usata nei due dataset è differente; in __books_ratings__ la scala di valutazione è da 0 a 10, mentre in __goodbooks__ è da 0 a 5. Perciò dividiamo per 2 i voti in __books_avg_rating__.

Per rendere più chiaro il risultato rinominiamo le colonne tramite la funzione _rename_.

```python
books_total = books_avg_rating.merge(goodbooks, left_on='ISBN', right_on='isbn')
books_total['Dif'] = books_total['average_rating']-books_total['Avg-Rating']/2
books_total.rename(columns={'Avg-Rating':'Books-Average-Rating','average_rating':'Goodbooks-Average-Rating'}, inplace=True)
books_total[['ISBN','Book-Title','Books-Average-Rating', 'Goodbooks-Average-Rating', 'Dif']].head()
```
